```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Neighborhood": [
        "Strip District",
        "Allentown",
        "Lincoln-Lemington-Belmar",
        "South Side Flats",
        "Greenfield",
        "Central Northside",
        "Brookline",
        "Arlington",
        "Hill District",
        "Carrick",
    ],
    "Visits": [
        1529,
        552,
        371,
        177,
        550,
        292,
        695,
        354,
        286,
        738,
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(df['Neighborhood'], df['Visits'], color='skyblue')
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Visits', fontsize=12)
plt.title('Community Center Visits by Neighborhood', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Show the graph
plt.show()
```
